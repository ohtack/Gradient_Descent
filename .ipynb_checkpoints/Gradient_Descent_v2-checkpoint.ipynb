```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

<b>Generate data</b>
\begin{equation} Y = 2X + 3 \end{equation}
\begin{equation} W = 2    \\b = 3\end{equation}

```python
X = np.random.rand(100)
y = 2 * X  + 3
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X,y)
plt.show()
```

```python
def plot_prediction_1 (pred, y):
    plt.figure(figsize=(10, 8))
    plt.scatter(X,y)
    plt.scatter(X,pred)
    plt.show()
```

<b>Generate random W and b and Gradient_Descent</b>

\begin{equation} \hat{y} = WX + b \end{equation}


<b>Cost Fuction</b>
\begin{equation}
L(X) = {\frac{1}{n}} \sum_{i=1}^{n} (\hat{y} - y_i)^2 
\end{equation}

<b> Gradient descent</b>

\begin{equation}
W(grad) = lr \times \frac{1}{n} \sum_{i=1}^{n} ((\hat{y} - y)X   \\
b(grad) = lr \times \frac{1}{n} \sum_{i=1}^{n} (\hat{y} - y)\end{equation}

```python
W = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate = 0.7

for epoch in range(200):
    y_pred = W * X + b
    error = np.abs(y_pred - y).mean()
    
    if error < 0.0005:
        break
        
    W_grad = learning_rate * ((y_pred - y)*X).mean()
    b_grad = learning_rate * (y_pred - y).mean()
    
    # W, b  value update
    
    W = W - W_grad
    b = b - b_grad
    
    if epoch % 10 == 0:
        print("Epoch = %d" %(epoch))
        print("W     = %0.03f" %(W))
        print("b     = %0.03f" %(b))
        print("error = %0.04f" %(error))
        y_pred = W * X + b
        plot_prediction_1(y_pred, y)
```

<b>Additional independent variable</b>


\begin{equation} Y = 2X_1 + 3X_2 + 1 \end{equation}

```python
X1 = np.random.rand(100)
X2 = np.random.rand(100)
y = 2 * X1 + 3 * X2  + 1
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X1, X2, y, marker='o', s=20, c = 'red')
```

```python
def plot_prediction_2 (pred, y):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    for c, m, i in [('r', 'o', y), ('b', '^', pred)]:
        ax.scatter(X1, X2, i, c=c, marker=m)
```

```python
W1 = np.random.uniform(-1,1)
W2 = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate = 0.7
```

```python
for epoch in range(200):
    
    y_pred = W1 * X1 + W2 * X2 + b
    
    error = np.abs(y_pred - y).mean()
    if error < 0.0005:
             
        break
    
    # gradient descent
    
    W1_grad = learning_rate * ((y_pred - y) * X1).mean()
    W2_grad = learning_rate * ((y_pred - y) * X2).mean()
    b_grad = learning_rate * (y_pred - y).mean()
    
    # W, b  value update
    W1 = W1 - W1_grad
    W2 = W2 - W2_grad
    b = b - b_grad
    
    if epoch % 10 == 0:
        print("Epoch = %d" %(epoch))
        print("W1    = %0.03f" %(W1))
        print("W2    = %0.03f" %(W2))
        print("b     = %0.03f" %(b))
        print("error = %0.04f" %(error))
        y_pred = W1 * X1 + W2 * X2 + b
        plot_prediction_2(y_pred, y)
    plt.show()
```
